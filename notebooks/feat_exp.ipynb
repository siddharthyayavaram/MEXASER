```python
mapping_dict = {}

for index, string in enumerate(smile.feature_names):
    key = string.split('_')[:-1]
    key = '_'.join(key)
    
    # If the key is not already in the dictionary, add it with its index
    if key not in mapping_dict:
        mapping_dict[key] = [index]
    else:
        mapping_dict[key].append(index)

print(mapping_dict)
```

```python
map = {}

for index, string in enumerate(smile.feature_names):
    key = string.split('_')[0]
    # key = '_'.join(key)
    
    # If the key is not already in the dictionary, add it with its index
    if key not in map:
        map[key] = [index]
    else:
        map[key].append(index)
```

```python
def find_key_md(x):
    for key, indices in mapping_dict.items():
        if x in indices:
            return key
    return "No key found for value {}".format(x)

def find_key_map(x):
    for key, indices in map.items():
        if x in indices:
            return key
    return "No key found for value {}".format(x)

x = 4153
print("Key for value {}: {}".format(x, find_key_md(x)))
print("Key for value {}: {}".format(x, find_key_map(x)))
```
